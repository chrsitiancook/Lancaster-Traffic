```python
import pandas as pd

import numpy as np

import folium

from folium.plugins import HeatMap

url = 'https://github.com/chrsitiancook/Lancaster-Traffic/blob/main/DATA%20SOURCES/LANCASTER_2021/CRASH_LANCASTER_2021.csv?raw=true'
df = pd.read_csv(url, index_col = 0)
```

```python
map_obj = folium.Map(location = [40.037804, -76.305764], zoom_start = 10)
```

```python
df.head()
```

```python
dow_replace = {1:'Sunday', 2:'Monday',3:'Tuesday',4:'Wednesday',5:'Thursday',6:'Friday',7:'Saturday',9:'Unknown'}

df.replace({'DAY_OF_WEEK': dow_replace}, regex=True, inplace=True)
```

```python
df['DAY_OF_WEEK'].value_counts()
```

```python
#df['LOC']= str(df['DEC_LAT']) + str(df['DEC_LONG'])
loc_list = []

lat_list = df['DEC_LAT'].to_list()
long_list = df['DEC_LONG'].to_list()

for i in range(len(lat_list)):
#for i in range(100):
    if not pd.isna(lat_list[i])  and not pd.isna(long_list[i]):
        loc_list.append([lat_list[i], long_list[i]])
       # print(type(lat_list[i]))
      

#print(loc_list)

```

```python
HeatMap(loc_list).add_to(map_obj)

map_obj
```
