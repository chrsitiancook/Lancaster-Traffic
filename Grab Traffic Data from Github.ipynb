```python
import pandas as pd

url = 'https://github.com/chrsitiancook/Lancaster-Traffic/blob/main/DATA%20SOURCES/LANCASTER_2021/CRASH_LANCASTER_2021.csv?raw=true'
df = pd.read_csv(url, index_col = 0)
```

```python
df.head()
```

```python
dow_replace = {1:'Sunday', 2:'Monday',3:'Tuesday',4:'Wednesday',5:'Thursday',6:'Friday',7:'Saturday',9:'Unknown'}

df.replace({'DAY_OF_WEEK': dow_replace}, regex=True, inplace=True)
```

```python
df['DAY_OF_WEEK'].value_counts()
```

```python
df['LOC']= str(df['DEC_LAT']) + str(df['DEC_LONG'])
```
